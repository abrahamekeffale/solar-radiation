```python
# Define a function to handle negative values
def handle_negative_values(df, columns, method='zero'):
    for column in columns:
        if method == 'zero':
            df[column] = df[column].apply(lambda x: max(x, 0))
        elif method == 'mean':
            mean_value = df[column].loc[df[column] >= 0].mean()
            df[column] = df[column].apply(lambda x: max(x, mean_value))
        elif method == 'median':
            median_value = df[column].loc[df[column] >= 0].median()
            df[column] = df[column].apply(lambda x: max(x, median_value))
    return df

# List of columns where negative values should be addressed
columns_to_fix = ['GHI', 'DNI', 'DHI', 'WS', 'WSgust']

# Apply the handling method (e.g., replacing negative values with zero)
df_cleaned = handle_negative_values(df, columns_to_fix, method='zero')

# Verify the changes
print("Number of rows after handling negative values:", df_cleaned.shape[0])
print("Number of negative values remaining:")
for column in columns_to_fix:
    print(f"{column}: {df_cleaned[df_cleaned[column] < 0].shape[0]}")
```

Adjust IQR Multiplier:
 adjust the multiplier from 1.5 to a higher value to be more lenient or stricter in defining outliers.

```python
# List of columns to check for outliers
columns_to_check = ['GHI', 'DNI', 'DHI']
def remove_outliers_iqr(df, columns, multiplier=1.5):
    df_cleaned = df.copy()
    for column in columns:
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    return df_cleaned

# Try a different multiplier
df_cleaned = remove_outliers_iqr(df, columns_to_check, multiplier=2.0)
```

Calculate Z-scores and remove outliers based on a Z-score threshold.

```python
from scipy.stats import zscore

def remove_outliers_zscore(df, columns, threshold=3):
    df_cleaned = df.copy()
    for column in columns:
        z_scores = np.abs(zscore(df_cleaned[column]))
        df_cleaned = df_cleaned[z_scores < threshold]
    return df_cleaned

# Apply Z-score method
df_cleaned = remove_outliers_zscore(df, columns_to_check)
```

Cap extreme values to reduce the effect of outliers.

```python
from scipy.stats.mstats import winsorize

def winsorize_data(df, columns, limits=[0.05, 0.05]):
    df_cleaned = df.copy()
    for column in columns:
        df_cleaned[column] = winsorize(df_cleaned[column], limits=limits)
    return df_cleaned

# Apply Winsorization
df_cleaned = winsorize_data(df, columns_to_check)
```
